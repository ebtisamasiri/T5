## Data Preprocessing
Do what you think you need such as:
* Remove the outliers
* Impute missing data
* Scale the data
* Reduce dimentions using PCA
* Implement One-Hot Encoding for nominal categorical variables.

```python
#1-Remove the outliers
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
```

```python
#2- Impute missing data
cols_to_fill = [
    'Accident_Severity',
    'Number_of_Casualties',
    'Urban_or_Rural_Area'
]
df[cols_to_fill] = df[cols_to_fill].fillna(df[cols_to_fill].mean())
# df['Light_Conditions'] = df['Light_Conditions'].fillna(df.mode(), inplace=True)

mode_val = df['Light_Conditions'].mode()[0]
df['Light_Conditions'].fillna(mode_val, inplace=True)
```

```python
df.isnull().sum()
```

```python
#3- Scale the data
from sklearn.preprocessing import StandardScaler, OneHotEncoder
df = pd.get_dummies(df, df.select_dtypes(include=[object]).columns, drop_first=True)

scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)
```

```python
#4- Reduce dimentions using PCA
from sklearn.decomposition import PCA

n_components = 2
pca = PCA(n_components=n_components)
pca.fit(df)
x_pca = pca.transform(df)

df_pca1 = pd.DataFrame(x_pca,
                       columns=['PC{}'.
                       format(i+1)
                        for i in range(n_components)])
print(df_pca1)
```
